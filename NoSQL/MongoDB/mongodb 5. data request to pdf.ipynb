use MongoDB's schema validation to enforce a specific schema on certain fields without having to specify validation rules for every field in the document.

Here's an example of how you might set up schema validation on a collection to enforce validation on a single field, while leaving all other fields unchecked:

```javascript
db.createCollection("yourCollection", {
  validator: {
    $jsonSchema: {...
    }
  }
});
```


You can also use the `collMod` command to modify an existing collection and add schema validation in a similar manner. Here's an example:

```javascript
db.runCommand({
  collMod: "yourCollection",
  validator: {
    $jsonSchema: {
      bsonType: "object",
      properties: {
        fieldToValidate: {
          bsonType: "string",
          description: "fieldToValidate must be a string and is required"
        }
      },
      required: ["fieldToValidate"]
    }
  }
});
```

This command modifies the existing `yourCollection`, where the `fieldToValidate` field must be a string, but other fields in the documents are not subject to validation. You can customize the validation rules for `fieldToValidate` according to your needs..

Keep in mind that schema validation can have implications for your application's behavior and performance, so it's wise to test thoroughly after implementing or modifying validation rules.

````python
# aggregate level calculation
# all metrics are added to the exchange and after submission by the companies shown on the dashboard
# each time, you'll have a metricValue, which has an absolute value (as detailed in its own calculation doc), a description, and then a weighted one
# here we'll be adding up weighted values 
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase.pdfmetrics import registerFontFamily


class PDFGenerator:
    def __init__(self):
        font_path = "C:\\Users\\johan\\Documents\\GitHub\\supernova-py\\api_orion\\static\\TTF_fonts\\Soehne-Kraeftig.ttf"
        pdfmetrics.registerFont(TTFont("Soehne-Kraeftig", font_path))
        registerFontFamily('Soehne-Kraeftig',normal='Soehne-Kraeftig') #,bold='VeraBd',italic='VeraIt',boldItalic='VeraBI') # this is essential to use <b></b> etc.

        # Set self.styles and assign fonts from the fontfamily
        self.styles = getSampleStyleSheet()
        
        # self.styles.add(ParagraphStyle(name='Soehne-Kraeftig',
        #                       parent=self.styles['Normal'],
        #                       fontName='Soehne-Kraeftig'))

        self.title_style = ParagraphStyle('normal_style', self.styles['Heading1'], fontName='Soehne-Kraeftig') # was set to simply: self.styles['Heading1']
        self.normal_style = ParagraphStyle('normal_style', self.styles['Normal'], fontName='Soehne-Kraeftig')
        self.title_style_c = ParagraphStyle('title_style', parent=self.styles['Heading1'], 
                                            alignment=1, fontName='Soehne-Kraeftig') 
        self.normal_style_c = ParagraphStyle('normal_style', parent=self.styles['Normal'], 
                                             alignment=1, fontName='Soehne-Kraeftig')
        self.sub_style = self.styles['Heading2']
        self.comp_style = ParagraphStyle("CustomStyle", parent=self.styles['Heading2'],
                                         leftIndent=20, fontName='Soehne-Kraeftig')
        self.custom_style = ParagraphStyle("CustomStyle",parent=self.styles['Normal'],
                                           leftIndent=20, fontName='Soehne-Kraeftig',leading=18)


    def aggregate_sum(self, metricMeta_id, exchange):
        sum = 0
        for comp in exchange['companies']:
            entitity_id = comp['company_link']
            if metricMeta_id in comp['requested_metrics']:
                    metricValue = db.metricValues.find_one({'company_link': entitity_id})
                    if metricValue:
                        sum += float(metricValue['value'])
        return sum
        
    def onFirstPage(self, canvas, document):
        # Draw your table on the first page
        table_data = [
            [Paragraph('Logo', self.title_style_c), Paragraph('Fund', self.normal_style_c)],
            [Paragraph('Aggregate', self.title_style_c), Paragraph('Period', self.normal_style_c)]
        ]
        
        table = Table(table_data, colWidths=[250, 390], rowHeights=80)
        table.setStyle([
            ('FONT', (0,0), (-1,-1), 'Soehne-Kraeftig'), 
            ('VALIGN', (0,0), (-1,-1), 'MIDDLE'),
            ('ALIGN', (0,0), (-1,-1), 'LEFT'),
            ('LEADING', (0,0), (-1,-1), 4, 'LEFT'),
            ('INNERGRID', (0,0), (-1,-1), 0.25, colors.black),
            # ('GRID', (0,0), (-1,-1), 1, colors.lightgrey),
            # ('LINEBELOW', (0, 0), (-1, -2), 1, colors.grey),  # Lines below all rows except last
            ('LINEBELOW', (0, -1), (-1, -1), 1, colors.grey),
            # ('LINEAFTER', (0, 0), (0, -1), 1, colors.grey)
        ])
        table.wrapOn(canvas, 300, 800)
        table.drawOn(canvas, 0, 690)
        
        # canvas.setStrokeColor(colors.grey)
        # # Draw horizontal line along the entire width
        # canvas.line(50, 750, A4[0] - 50, 750)
        # # Draw vertical line along the entire height
        # canvas.line(A4[0]/2, 50, A4[0]/2, 800)

    def onLaterPages(self, canvas, document):
        # Draw a header on later pages
        header = Table([[Paragraph('Logo', self.title_style_c), Paragraph('Fund', self.normal_style_c)]], colWidths=[250, 390], rowHeights=80)
        header.setStyle([
            ('FONT', (0,0), (-1,-1), 'Helvetica'), 
            ('VALIGN', (0,0), (-1,-1), 'MIDDLE'),
            ('ALIGN', (0,0), (-1,-1), 'CENTER'),
            ('INNERGRID', (0,0), (-1,-1), 0.25, colors.black),
            ('LINEBELOW', (0, -1), (-1, -1), 1, colors.grey),
        ])
        header.wrapOn(canvas, 300, 400)
        header.drawOn(canvas, 0, 765)
        
    def aggregate(self, metricMeta_id, exchange_id): # exchange_id??
        
        # find the request
        exchange:dict = db.exchange.find_one({'_id': ObjectId(exchange_id)})

        # find the metricMeta
        metricMeta:dict = db.metricMeta.find_one({'_id': ObjectId(metricMeta_id)}, {'embeddings':0})
        
        # logging.info("It found the request %s and associated fund %s on mongodb", str(request_id), str(entitity_id))
        
        # Create a new document
        # buffer = BytesIO()
        doc = SimpleDocTemplate('test.pdf', pagesize=A4, topMargin=80)
        
        # logo_path = os.path.join(os.path.dirname(__file__), 'static', 'supernova-logo.png')
        # logo = Image(logo_path, width=105, height=25) 
        # logo = Paragraph('logo', self.title_style_c)

        # # set heading
        # title = Paragraph('Aggregate', self.title_style_c)
        # fund = Paragraph(f"In fund: {exchange['fund_link']}", self.normal_style_c)
        # period = Paragraph(f"Reporting period: {' - '.join(exchange['reporting_period'])}<br/>Deadline: {exchange['deadline']}", self.normal_style_c)

        # # Create tables
        # table_data = [
        #     [logo, fund],
        #     [title, period]]

        # table = Table(table_data) # colWidths=[150, 290], rowHeights=80
        # table.setStyle([('FONT', (0,0), (-1,-1), 'Soehne-Kraeftig'), # SÃ¶hne
        #                 ('VALIGN', (0,0), (-1,-1), 'MIDDLE'),
        #                 ('ALIGN', (0,0), (-1,-1), 'CENTER'),
        #                 ('GRID', (0,0), (-1,-1), 1, colors.lightgrey)]) 
        
        # Create a list to hold the elements to add to the document
        elements: list[Flowable] = [Spacer(1, 90)] #table]
        
        # aggregate # unit !?
        sum = generatepdf.aggregate_sum(metricMeta_id, exchange)
                        
        elements.append(Spacer(1, 15))                
        elements.append(Paragraph(f"{sum} units were delivered", self.title_style_c))
        
        elements.append(Spacer(1, 5))
        
        horizontal_line = HRFlowable(width="100%", thickness=1, lineCap='round', color=colors.lightgrey)
        elements.append(horizontal_line)
        
        # indicator description
        elements.append(Spacer(1, 5))
        elements.append(Paragraph('INDICATOR', self.sub_style))
        elements.append(Paragraph(metricMeta['title'], self.comp_style))
        elements.append(Spacer(1, 5))
        elements.append(Paragraph(metricMeta['description'], self.normal_style))
        elements.append(Spacer(1, 5))
        elements.append(Paragraph(metricMeta['Usage Guidance'], self.normal_style))
        elements.append(Spacer(1, 5))
        if 'Calculation' in metricMeta.keys():
            elements.append(Paragraph(metricMeta['Calculation'], self.normal_style))
            elements.append(Spacer(1, 5))

        elements.append(Paragraph(metricMeta['Footnote'], self.normal_style))
        elements.append(Spacer(1, 5))
        elements.append(Paragraph('''<u><link href="http://www.example.com">Read more on the metric</link></u>''', self.custom_style))
        
        # companies
        elements.append(Spacer(1, 15))  
        elements.append(horizontal_line)
        elements.append(Spacer(1, 5))
        elements.append(Paragraph('COMPANIES', self.sub_style))
        elements.append(Spacer(1, 5))

        # loop over the companies in the request
        for comp in exchange['companies']:
            entitity_id = comp['company_link']
            company:dict = db.companies.find_one({'_id': ObjectId(entitity_id)})
            if company.get('title', {}):
                elements.append(Paragraph(company['title'], self.comp_style))
                
                # if the company in the request has the metric, take the value
                if metricMeta_id in comp['requested_metrics']:
                    metricValue:dict = db.metricValues.find_one({'company_link': entitity_id})

                    if metricValue:
                        # metricMeta = db.metricValues.find_one({"_id": metricValue['metric_meta_link']})
                        companyValue = 'Value: <b>' + str(metricValue['value']) + "</b> units"
                        period = "For the reporting period: " + ' - '.join(metricValue['reporting_period']) # + metricMeta['Unit'] !!!
                        
                    else: 
                        companyValue = "No value"
                                    
                    elements.append(Paragraph(f"{companyValue} <br/> {period} <br/> Explanation of calculation method (coming soon)<br/><u><link href='http://www.example.com'>Read more on entity-level calculation here</link></u>", self.custom_style))
            else: continue    
            elements.append(Spacer(1, 15)) 
            elements.append(horizontal_line)
        
        # Build the PDF
        return doc.build(elements, onFirstPage=self.onFirstPage, onLaterPages=self.onLaterPages)

generatepdf = PDFGenerator()
generatepdf.aggregate("64da2bcae5031249c4c7fc87", "64edf0fbd7b8e9a7f63d1652")
````
