# On noise

Adding noise to a sample can be done in many ways.

```python
mu, sigma = 0, 0.1 # mean and standard deviation

s = np.random.normal(mu, sigma, (1000, 5))  # X.shape = 1000, 5
plt.hist(s)                                 # see: ML 2.2 KNN classification.ipynb
```

You can do the same thing without specifying $\mu$ and $\sigma$. But not that the amount of noise then becomes +/- 3.

```python
s = np.random.normal(size=(1000, 5))  # X.shape = 1000, 5
plt.hist(s)   
```

This is essentially the same as `randn()`, used in [this notebook (path)](</C:\Users\johan\OneDrive\/_Msc Data Science\statistics\MissingData-Correlation.ipynb>)

```python
s = np.random.randn(1000)
plt.hist(s)
```
