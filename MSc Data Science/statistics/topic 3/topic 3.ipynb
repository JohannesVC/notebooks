```python
sample = stats.poisson.rvs(2, size=100)

_, axs = plt.subplots(1,2, figsize=(15,5))

s = pd.Series(sample)
axs[0] = s.plot.hist(density=True, xlim=[0,13], ax=axs[0])
s.plot.density(ax=axs[0])

axs[1] = s.interpolate(method='linear').plot.hist(ax=axs[1], density=True) # 
s.interpolate(method='linear').plot.density(ax=axs[1])

plt.show()
```
