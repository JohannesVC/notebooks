## Power laws and TF-IDF

### 1. Power law

```python
# How a scale_factor of 2 affects the power law but not its scale invariance
original_a = 1.3
k = 0.25
scale_factor = 2**k 

# New coefficient
new_a = original_a * scale_factor

x_values = np.linspace(0.1, 10, 400)
original_y = original_a * x_values**k
scaled_y = new_a * x_values**k

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_values, original_y, label=f'Original Y = {original_a}X^{{{k}}}', color='blue')
plt.plot(x_values, scaled_y, label=f'Scaled Y = {new_a:.4f}X^{{{k}}}', color='red', linestyle='--')
plt.title('Comparison of Original and Scaled Power Law Functions')

plt.xlabel('X');plt.ylabel('Y');plt.legend();plt.grid(True);plt.show()
```

```python
# On a log scale
plt.figure(figsize=(12, 8))
plt.plot(x_values, original_y, label=f'Original Y = {original_a}X^{{{k}}}', color='blue')
plt.plot(x_values, scaled_y, label=f'Scaled Y = {new_a:.4f}X^{{{k}}}', color='red', linestyle='--')

plt.gca().set_xscale('log')
plt.gca().set_yscale('log')

plt.title('Effects of Different Coefficients and Exponents in Power Laws')
plt.xlabel('x (log scale)')
plt.ylabel('y (log scale)')
plt.legend()
plt.grid(True, which="both", ls="--", color='gray')
plt.show()
```

### 2. TF-IDF
See word file. Note that the base of the log can vary. I plotted it below for reference.
 
 
$$ {TF}_{ij}\ =\frac{f_{ij}}{{max}_{k\ \cdot}\ f_{kj}}  \text{ and } IDF_i={log}_2\left(\frac{N}{n_i}\right) $$
---

*Exercise 1.3.1: Suppose there is a repository of ten million documents. What (to the nearest integer) is the IDF for a word that appears in (a) 40 documents (b) 10,000 documents?*
Applying the IDF formula above:

$IDF_a=\log_2{\left(\frac{10000000}{40}\right)}=\log_2{\left(25000\right)\approx\mathbf{15}\ }$
$IDF_b=\log_2{\left(\frac{10000000}{10000}\right)}=\log_2{\left(100\right)\approx\mathbf{7}\ }\  $

---
*Exercise 1.3.2: Suppose there is a repository of ten million documents, and word w appears in 320 of them. In a particular document d, the maximum number of occurrences of a word is 15. Approximately what is the TF.IDF score for w if that word appears (a) once (b) five times?*

$ IDF_c=\log_2{\left(\frac{100000000}{320}\right)}=\ 15 $

The times word a appears is one. So, $f_{ij}=1$.  

The maximum number of occurrences of a word is 15. 

So, $max_k=15.$

$TF_{waj}=1/15$

So, the TF.IDF score for word a in document j is:
> $TF.IDF_{aj}=15\cdot1/15\ =\ \mathbf{1}$

The times word b appears is five. So, 
$TF_{wbj}=5/15\ =\ 3$

So, the TF.IDF score for word b in document j is:
> $TF.IDF_{aj}=3\cdot1/15\ =\mathbf{0}.\mathbf{2}$

### Notes on $logs$

A logarithm answers the question: **"To what power must the base \( b \) be raised, to produce a certain number \( N \)?"**. 

Mathematically, this is written as:

$\log_b(N) = x \quad \text{if and only if} \quad b^x = N$

Where:
- \( b \) is the base of the logarithm.
- \( N \) is the number you are taking the logarithm of.
- \( x \) is the power to which the base must be raised to equal \( N \).

Logarithms are used to simplify calculations in mathematics, especially when dealing with multiplication and division of large numbers, since they convert multiplication into addition and division into subtraction. In more practical terms, logarithms describe processes or phenomena that scale multiplicatively or exponentially, like radioactive decay, sound intensity, and certain financial calculations.

**Logarithmic Scaling** is seen in phenomena like the Richter scale for earthquakes, where each step represents a tenfold increase in measured amplitude and roughly 31.6 times more energy release.

In short, logs are most interesting for plotting purposes, where many phenomena in nature are of an exponential nature.
