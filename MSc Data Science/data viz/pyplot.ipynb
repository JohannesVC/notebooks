```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

From Topic 2 lesson 6.

2 APIs
Object-oriented vs big flat namespace
- pyplot is a large flat namespace and is recommended for exploratory first analysis
- it has a current figure object `plt` that makes it easy to work with - see `gcf()`
- mpl is object oriented and is recommended for customised plots
- mpl returns axes objects instead of the current figure in `plt`

```python
df = pd.DataFrame({
    'x': np.random.normal(0, 1, 100),   # mean, std, shape
    'y': np.random.normal(0, 2, 100)
})
```

```python
plt.hist(df)
plt.show()
```

```python
plt.hist(df.x, color='r')

plt.hist(df.y)

plt.show()
```

```python
plt.figure(figsize=(20,5))

plt.subplot(131)
plt.hist(df.x)
plt.xlim(-6,6)

plt.subplot(132)
plt.hist(df.y)
plt.xlim(-6,6)
plt.show()
```

### Now with the OO approach.

```python
fig, axs = plt.subplots(1,2, figsize=(20,8)
                        , sharex=True, sharey=True  # this gets rid of the y ticks 
                        )

axs[0].hist(df.x)
axs[0].set_xlim(-6,6)
axs[0].set_ylim(0,25)
axs[0].set_title("dist of x")

axs[1].hist(df.y)
axs[1].set_xlim(-6,6)
axs[1].set_title("dist of y")
axs[1].set_ylim(0,25)

plt.show()
```
